# **Reading video from webcam and displaying it in grayscale**

```
from google.colab.patches import cv2_imshow

import numpy as np
video =cv2.VideoCapture(0)
while(True):
    ret, frame = video.read()
    grayvideo = cv2.cvtColor(video, cv2.COLOR_BGR2GRAY)
    cv2.imshow('FinalVideo', grayvideo)
    cv2.waitKey(1)
    if cv2.waitKey(33)==27:
        break
cv2.destroyAllWindows()
```

```
video.release()
```

# **Creating white image using np.ones and draw line, rectangle, circle on it**

```
import numpy as np
```

```
picture = np.ones((400,400,3), dtype='uint8')*255

cv2_imshow(picture)
```

```
copyOfImage1 = picture.copy() 

line1 = cv2.line(copyOfImage1,(50,50),(350,350),(0,255,255),5)
line2 = cv2.line(copyOfImage1,(50,350),(350,50),(0,0,255),5)
line3 = cv2.line(copyOfImage1,(200,50),(200,350),(255,0,255),5)
line4= cv2.line(copyOfImage1,(100,200),(300,200),(0,255,255),5)

cv2_imshow(copyOfImage1)

```

```
copyOfImage2=picture.copy()
rectangle1 = cv2.rectangle(copyOfImage2,(50,50),(350,350),(0,255,255),10)
rectangle2 = cv2.rectangle(copyOfImage2,(100,150),(300,300),(255,0,0),3)

cv2_imshow(copyOfImage2)
```

```
copyOfImage3=picture.copy()

(x,y)=(int(picture.shape[1]/2), int(picture.shape[0]/2))
circle=cv2.circle(copyOfImage3, (x,y), (120), (0,0,255),3)

cv2_imshow(circle)
```

# **Drawing eclipse and polygon**

```
copyOfImage4 = picture.copy()

(x, y) = (int(picture.shape[1]/2), int(picture.shape[0]/2))
cv2.ellipse(copyOfImage4,(x,y),(150,50),0,0,360,(0,255,250),-1)
cv2_imshow(copyOfImage4)
```

```
copyOfImage5 = picture.copy()
pts = np.array([[100,50],[200,300],[370,200],[150,100]], np.int32)
pts = pts.reshape((-1,1,2))
polygon=cv2.polylines(copyOfImage5,[pts],True,(0,0,255),3)
cv2_imshow(copyOfImage5)
```
